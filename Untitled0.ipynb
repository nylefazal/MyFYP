```python
import re
from googleapiclient.discovery import build
import ipywidgets as widgets
from IPython.display import display, clear_output
from youtube_transcript_api import YouTubeTranscriptApi

# Function to extract YouTube video ID
def extract_video_id(url):
    video_id = re.search(r"v=([^\&]+)", url)
    return video_id.group(1) if video_id else None

# YouTube Data API function to get video details
def get_video_details(video_id, api_key):
    youtube = build('youtube', 'v3', developerKey=api_key)

    request = youtube.videos().list(
        part="snippet,statistics",
        id=video_id
    )
    response = request.execute()

    if response['items']:
        video_info = response['items'][0]
        title = video_info['snippet']['title']
        views = video_info['statistics']['viewCount']
        description = video_info['snippet']['description']
        return title, views, description
    return None

# Function to fetch video transcript
def get_video_transcript(video_id):
    try:
        transcript = YouTubeTranscriptApi.get_transcript(video_id)
        formatted_transcript = ""
        last_speaker = None

        for t in transcript:
            # Here, we can just format the output, since the API usually does not provide speaker labels.
            text = t['text'].strip()
            # For demonstration, let's alternate speakers (this part can be more complex based on actual speaker data)
            if last_speaker is None or last_speaker == "Speaker 1":
                formatted_transcript += f"**Speaker 1:** {text}\n"
                last_speaker = "Speaker 2"
            else:
                formatted_transcript += f"**Speaker 2:** {text}\n"
                last_speaker = "Speaker 1"

        return formatted_transcript
    except Exception as e:
        return str(e)

# Function to fetch and display video details
def fetch_video_info(button):
    clear_output()  # Clear previous outputs
    display(url_input, get_details_button, output_text)  # Redisplay widgets

    url = url_input.value
    video_id = extract_video_id(url)

    if video_id:
        api_key = "AIzaSyC_WCf-aMwPwHP3KRDHV-gKvMZbhce26_k"  # Your API key
        details = get_video_details(video_id, api_key)

        if details:
            title, views, description = details

            # Fetch transcript
            transcript = get_video_transcript(video_id)

            # Format the output for better readability
            output_text.value = (
                f"üé• **Title:** {title}\n\n"
                f"üëÄ **Views:** {views}\n\n"
                f"üìú **Description:**\n{description}\n\n"
                f"üìù **Transcript:**\n{transcript}"
            )
        else:
            output_text.value = "üö´ Could not retrieve video details."
    else:
        output_text.value = "‚ùå Invalid YouTube URL."

# Create the GUI
url_input = widgets.Text(
    description='YouTube URL:',
    placeholder='Enter YouTube video link'
)
get_details_button = widgets.Button(description="Get Video Details")
output_text = widgets.Textarea(
    description='Output:',
    layout=widgets.Layout(width='100%', height='400px'),
    disabled=True
)

# Link the button to the function
get_details_button.on_click(fetch_video_info)

# Display the widgets
display(url_input, get_details_button, output_text)
```
